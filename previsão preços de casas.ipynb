```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
x=np.arange(0,20,1).reshape(-1,1)
y=np.arange(0,40,2).reshape(-1,1)
noise = np.random.normal(0,10,len(x))
y = [elemA + elemB for elemA, elemB in zip(y,noise)]
plt.scatter(x,y)
```

```python
scaler_x=StandardScaler()
x=scaler_x.fit_transform(x)

scaler_y=StandardScaler()
y=scaler_y.fit_transform(y)

b0 = tf.Variable(0.54)
b1 = tf.Variable(0.71)
```

```python
a=tf.Variable(0.1)#termo dependente
b=tf.Variable(0.1)#termo independente
erro=lambda: tf.losses.mean_squared_error(y,(x*a+b))
otimizador=tf.keras.optimizers.SGD(learning_rate=0.0001)

```

```python

for i in range(5000):
     treinamento = otimizador.minimize(erro,var_list=[a,b])
    

    
```
